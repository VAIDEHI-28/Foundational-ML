```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(11 , 8))
sns.heatmap(df.corr() , annot = True)
```

```python
df['quality'].value_counts()
df['quality'].value_counts().plot(kind = 'bar')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Select the columns you want to plot
columns_to_plot = df.columns[1:11]  # First 11 columns, adjust as needed

# Setup Subplots (4 rows, 3 columns to fit 11 plots)
fig, axes = plt.subplots(4, 3, figsize=(15, 12))  # 4x3 grid (12 spots, 11 used)

axes = axes.flatten()  # Flatten to make indexing easier

# Plot each column's histogram
for i, col in enumerate(columns_to_plot):
    sns.histplot(df[col], kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Histogram of {col}')

# Hide any unused subplot (last one will be empty)
for j in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Select all columns you want to plot
columns_to_plot = df.columns

# Setup Subplots (adjust grid size based on the number of columns)
n_cols = len(columns_to_plot)
n_rows = (n_cols + 2) // 3 # Roughly 3 columns per row

fig, axes = plt.subplots(n_rows, 3, figsize=(15, n_rows * 4))

axes = axes.flatten()  # Flatten to make indexing easier

# Plot each column's histogram
for i, col in enumerate(columns_to_plot):
    sns.histplot(df[col], kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Histogram of {col}')

# Hide any unused subplots
for j in range(n_cols, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
