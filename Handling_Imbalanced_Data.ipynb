```python
import numpy as np
import pandas as pd

#Set the random seed for reproducibility
np.random.seed(123)

#Create a df with two samples
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0 , n_class_1
```

```python
#Cretae a df with imbalanced data

class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0 , scale=1 , size=n_class_0),
    'feature_2' : np.random.normal(loc=0 , scale=1 , size=n_class_0),
    'target' : [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2 , scale=1 , size=n_class_1),
    'feature_2' : np.random.normal(loc=2 , scale=1 , size=n_class_1),
    'target' : [1] * n_class_1
})
```

```python
df = pd.concat([class_0 , class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority , replace = True , n_samples = len(df_majority) , random_state = 42)
```

```python
df_upsampled = pd.concat([df_majority , df_minority_upsampled])
df_upsampled
```

```python
#DOWN SAMPLING

import numpy as np
import pandas as pd

#Set the random seed for reproducibility
np.random.seed(123)

#Create a df with two samples
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

n_class_0 , n_class_1

#Cretae a df with imbalanced data

class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0 , scale=1 , size=n_class_0),
    'feature_2' : np.random.normal(loc=0 , scale=1 , size=n_class_0),
    'target' : [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2 , scale=1 , size=n_class_1),
    'feature_2' : np.random.normal(loc=2 , scale=1 , size=n_class_1),
    'target' : [1] * n_class_1
})
df = pd.concat([class_0 , class_1]).reset_index(drop=True)
df.head()
```

```python
df['target'].value_counts()
```

```python
#downsampling

df_majority = df[df['target']==0]
df_minority = df[df['target']==1]
```

```python
from sklearn.utils import resample
df_majority_downsampled = resample(df_majority , replace = False , n_samples = len(df_minority) , random_state = 42)
```

```python
df_downsampled = pd.concat([df_majority_downsampled , df_minority])
```

```python
df_downsampled['target'].value_counts()
```

```python
df_downsampled.head()
```

```python
df_downsampled.tail()
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

majority = pd.DataFrame({
    'Feature1': np.random.normal(0, 1, 950),
    'Feature2': np.random.normal(0, 1, 950),
    'Class': 0
})

minority = pd.DataFrame({
    'Feature1': np.random.normal(2, 1, 50),
    'Feature2': np.random.normal(2, 1, 50),
    'Class': 1
})

df = pd.concat([majority, minority], ignore_index=True)

df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print(df['Class'].value_counts())
print(df.head())
```

```python
df['Class'].value_counts()
```

```python
df_majority = df[df['Class']==0]
df_minority = df[df['Class']==1]
```

```python
#upsampling

from sklearn.utils import resample
df_minority_upsampled = resample(df_minority , replace = True , n_samples = len(df_majority) , random_state = 42)
```

```python
df_upsampled = pd.concat([df_majority , df_minority_upsampled])
```

```python
df_upsampled['Class'].value_counts()
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

majority = pd.DataFrame({
    'Feature1': np.random.normal(0, 1, 950),
    'Feature2': np.random.normal(0, 1, 950),
    'Class': 0
})

minority = pd.DataFrame({
    'Feature1': np.random.normal(2, 1, 50),
    'Feature2': np.random.normal(2, 1, 50),
    'Class': 1
})

df = pd.concat([majority, minority], ignore_index=True)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle

print(df['Class'].value_counts())
```

```python
sns.countplot(x='Class', data=df)
plt.title("Before Applying SMOTE")
plt.show()
```

```python
minimum , Q1 , median , Q3 , maxium
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
print(lower_fence , higher_fence)
```
