```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.drop('Route' , axis = 1 , inplace = True)
df.head()
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.drop('Date_of_Journey' , axis = 1 , inplace = True)
df.head(2)
```

```python
df['Date'] = df["Date"].astype(int)
df['Month'] = df["Month"].astype(int)
df['Year'] = df["Year"].astype(int)
```

```python
df.info()
```

```python
df['Duration'] = df['Duration'].astype(str)

df['Duration_hour'] = df['Duration'].str.extract(r'(\d+)h')[0].fillna(0).astype(int)

df['Duration_min'] = df['Duration'].str.extract(r'(\d+)m')[0].fillna(0).astype(int)

df[['Duration', 'Duration_hour', 'Duration_min']].head()


```

```python
df['Duration_hour'].unique()
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Duration' , axis = 1 , inplace = True)
df.head(2)
```

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df["Departure_min"] = df["Dep_Time"].str.split(':').str[1]
df.head(2)
```

```python
df.drop('Dep_Time' , axis = 1 , inplace=True)
df.head(2)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])

df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df.drop('Arrival_Time' , axis = 1 , inplace = True)
df.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
x = OneHotEncoder()

encoded = x.fit_transform(df[['Airline' , 'Source' , 'Destination']]).toarray()
```

```python
x_df = pd.DataFrame(x.fit_transform(df[['Airline' , 'Source' , 'Destination']]).toarray() , columns = x.get_feature_names_out())
x_df
```

```python
df.drop(['Airline' , 'Source' , 'Destination'] , axis = 1 , inplace = True)
df.head(2)
```

```python
df = pd.concat([df , x_df] , axis =1)
df.head()
```

```python
# Step 1: Extract the 'Price' column
price_column = df['Price']

# Step 2: Drop the 'Price' column from the DataFrame
df = df.drop('Price', axis=1)

# Step 3: Concatenate 'Price' at the end
df = pd.concat([df, price_column], axis=1)

# Step 4: Check your updated DataFrame
df.head()
```

```python
#Get all the col insted od addition_info
columns = [col for col in df.columns if col != 'Additional_Info']

#new order
new_order = columns[:-1] + ['Additional_Info'] + columns[-1:]

df = df[new_order]
df.head()
```
